## Feature Encoding

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import preprocessing
```

```python
df = pd.read_csv('Churn_Modelling.csv')
```

```python
# Get rid of irrelevant columns
df.drop(columns = ["CustomerId", "RowNumber", "Surname"], axis = 1, inplace=True)
```

```python
df.Gender.value_counts()
```

#### Handling missing values in Gender

```python
# find the mode
df.Gender.mode()

# fill the missing values with mode
df['Gender'] = df['Gender'].fillna('Male')

# see the male stats increase
df.Gender.value_counts()
```

```python
df.info()
```

#### Label Encoding

```python
# Give 'Male' a 0 and 'Female' a 1; or locations a 1,2,3... even though this is not recommended.

# sklearn preprocessor to Label Encoder. only used for Y variables

le = preprocessing.LabelEncoder()
df['Gender_label'] = le.fit_transform(df.Gender.values)
```

```python
# see a new created column of Gender_label
df.head(10)
```

#### Target Encoding

```python
pip install category_encoders
```

```python
from category_encoders import TargetEncoder

encoder = TargetEncoder()
```

```python
df1 = pd.read_csv('Churn_Modelling.csv')
df1.drop(columns = ["CustomerId", "RowNumber", "Surname"], axis = 1, inplace=True)
df1['Gender'] = df1['Gender'].fillna('Male')
```

```python
df1.head()
```

```python
df1['Gender Encoded'] = encoder.fit_transform(df1['Gender'], df1['Exited'])
```

```python
df1.head()
```

```python
df1['Gender Encoded'].value_counts()
```

#### Benefits of TE:

- Target encoding is a simple and quick encoding method that doesn't add to the dimensionality of the dataset. Therefore, it may be used as a good first try technique

#### Limitations of TE:

- It is dependent on the target distribution which means target encoding requires careful validation as it can be prone to overfitting.
